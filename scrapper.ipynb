```python
import re
import requests
from bs4 import BeautifulSoup
```

```python
url="https://www.rockriverarms.com/index.cfm?fuseaction=page.display&page_id=25"
r=requests.get(url)
```

```python
soup = BeautifulSoup(r.content, 'html.parser') 
soup1=soup.find_all('script')
```

```python
soup2=str(soup1[8])
soup2
```

```python
email = re.findall(r'[\w\.-]+@[\w\.-]+', soup2)
email
```

```python
company_name = re.findall(r'<strong>(.*?)</strong>', soup2)
len(company_name)
```

```python
website_name = re.findall(r'website:&nbsp;(.*?)<br>', soup2)
len(website_name)
```

```python
address_pattern =r'<br>([A-Za-z\s]+),&nbsp;[A-Z]{2}&nbsp;'

matches = re.findall(address_pattern, soup2)

len(matches)
```
